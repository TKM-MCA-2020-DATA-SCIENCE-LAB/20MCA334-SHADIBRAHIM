```python
import matplotlib.pyplot as plt
```

```python
import numpy as pd
x=pd.linspace(0,5,11)
y=x**2
print(x)
print(y)
```

```python
plt.plot(x,y,'g')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Variation")
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.subplot(1,3,2)
plt.plot(x,y,'g--')
plt.subplot(1,3,3)
plt.plot(x,y,'b--')
plt.show()
```

```python
#using object oriented method
fig=plt.figure()
axes=fig.add_axes([1,1,1,1])
axes.plot(x,y,'b')
axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")
axes.set_title("Title")
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.3,0.3])
axes1.plot(x,y,'r')
axes1.set_xlabel("X axis")
axes1.set_ylabel("Y axis")
axes1.set_title("first graph")
axes2.plot(x,y,'g')
axes2.set_xlabel("x axis")
axes2.set_ylabel("y axis")
axes2.set_title("second graph")

```

```python
fig,axes=plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")

```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
fig.tight_layout()
```

```python
for ax in axes:
  ax.plot(x,y,'b')
  axes2.set_xlabel("x axis")
  ax.set_ylabel("y axis")
  ax.set_title("second graph")
fig
```
